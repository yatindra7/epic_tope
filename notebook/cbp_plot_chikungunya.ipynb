```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/cbp/chikungunya.cstmidx.csv")
dataframe = df.drop(['Unnamed: 0'], axis=1)
print(dataframe.head())
print(dataframe.isnull().values.any())
```

```python
dataframe.describe()
```

```python
ax = sns.boxplot(data = dataframe)
ax.set_xticklabels(ax.get_xticklabels(), fontsize=8, rotation=45, ha="right")
plt.show()
```

```python
corr_matrix = dataframe.corr()

# plot the correlation matrix as a heatmap
plt.figure(figsize=(20, 16))
sns.heatmap(corr_matrix, cmap='coolwarm', annot=True)

# display the heatmap
plt.show()
```

```python
corr_pairs = corr_matrix.unstack()

# sort the series by correlation values and print the top 5 pairs
print(corr_pairs.sort_values())
```
